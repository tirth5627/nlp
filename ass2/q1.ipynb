```python
from visual_automata.fa.dfa import DFA
acc_dic={c:'q0' for c in set('abcdefghijklmnopqrstuvwxyz')}
reject_dic={c:'q_reject' for c in set('0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ !@#$%^&*()-_=+[]{}|;:\'",.<>?/`~')}
acc_dic.update(reject_dic)
dfa = DFA(
    states={'q0','q_reject'},
    input_symbols=set('abcdefghijklmnopqrstuvwxyz0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ !@#$%^&*()-_=+[]{}|;:\'",.<>?/`~'),
    transitions={
        'q0': acc_dic,
        'q_reject': {c: 'q_reject' for c in set('abcdefghijklmnopqrstuvwxyz0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ !@#$%^&*()-_=+[]{}|;:\'",.<>?/`~')}
    },
    initial_state='q0',
    final_states={'q0'}
)
dfa.show_diagram()
```
